#Part 4: Simple Machine learning questions

1.	Predict “How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form_and_Completed_and_submited_the_form?” for the complete year of 2022


```python
#Preprocessing of data, start with label encoding
colname=[]
for i in df.columns:
    if df[i].dtype=='object':
        colname.append(i)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in colname:
    df[i]=le.fit_transform(df[i])

df.head()

```

```python
#all data now in structure form
df=df.drop(["new_column"],axis=1) #drop unnecessary column
```

```python
#combine month and year column
df['Period'] = df['Month'].map(str) + '-' + df['Year'].map(str)
df.head()
```

```python
df=df.sort_values('Period', ascending = True)
df
```

```python
# reset index
data=data.set_index('Period')
```

```python
y=data['How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form_and_Completed_and_submited_the_form?']
```

```python
#split data to y
import matplotlib.pyplot as plt

y.plot()
plt.show()
```

```python
#for see trend and sesonality 
import statsmodels.api as sm
d=sm.tsa.seasonal_decompose(y)
fig = d.plot()
plt.show
```

```python
#according to above plot, data is negative trend and seasonality is present in between jan to july data
# so we use ARIMA model for predication
#do train test split
train= y.loc[:'2021-02-01']
test = y.loc['2021-03-01':]
```

```python
train.shape
```

```python
test.shape
```

```python
#ARIMA is not good for sesonal data, but we use holt-winter model for better prediction
from statsmodels.tsa.api import ExponentialSmoothing
Holt_Winter_df = test.copy()
fit1 = ExponentialSmoothing(train, seasonal_periods=4,trend = 'add', seasonal = 'add').fit()
Holt_Winter_df['Holt_Winter'] = fit1.forecast(steps=len(test))

train.plot(figsize=(10,6), title = 'N0. form', fontsize = 14)
test.plot(figsize =(10,6), title = 'No. form', fontsize = 14)
Holt_Winter_df['Holt_Winter'].plot(figsize=(10,6), title = 'No. form', fontsize = 14)
plt.show()
#blue color line is for train data
#orange colour line is for test data
#green colour line is for predicate data
```

2.	Generate the overall MAPE of your prediction for the year 2021

```python
mape = np.mean(np.abs((test - Holt_Winter_df['Holt_Winter'])/test))*100
print('Mean Absolute Percentage Error',mape)
print('Accuracy ',100-mape)
```

```python
#try Sarimax model
#find best model parameter
!pip install pmdarima
import pmdarima as pm
model = pm.auto_arima(y, start_p=0, max_p=3, d = None, max_d=2, start_q=0, max_q=3, satrt_P=3,D=None, max_D=2, start_Q=0,max_Q=3, max_order=10, m=4, seasonal=True, information_criterion = 'aic',test='adf', trace=True, random_state=10)
```

```python
model
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(y, order = (1,1,0), seasonal_order=(1,1,0,4)).fit()
```

```python
pred = model.get_prediction(start = pd.to_datetime('2021-03-01'))
plt.figure(figsize=(10,6))
ax=y['2019-01-01':].plot(label = "observed")
pred.predicted_mean.plot(ax=ax, label = 'validation')
ax.set_xlabel('date')
ax.set_ylabel('no.of form')
plt.legend()
plt.show()
```

```python
mape = np.mean(np.abs((test - pred.predicted_mean)/test))*100
print('Mean Absolute Percentage Error',mape)
print('Accuracy ',100-mape)
```

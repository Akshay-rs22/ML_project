# Missing Value Analysis

```python
#some variable contain "unknown" value so we should change it.
bank_df=bank_df.replace('unknown',np.nan)
```

```python
bank_df
```

```python
# in place of object datatype "unknown" value is replace with mode of same column.
# in place of int and float datatype "unknown" value is replace with mean of same column.

for x in bank_df.columns:
    if bank_df[x].dtype=='object' or bank_df[x].dtype=='bool':
        bank_df[x].fillna(bank_df[x].mode()[0],inplace=True) # some time mode is bimodal there for we want to select 1st mode use mode[0]
    elif bank_df[x].dtype=='int64' or bank_df[x].dtype=='float64':
        bank_df[x].fillna(round(bank_df[x].mean()),inplace=True)
```

#  Logistic Regression

```python
#model building for logistic regression.
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
Y_pred=classifier.predict(X_test)
print(Y_pred)
```

```python
print(list(zip(Y_test, Y_pred)))
```

```python
# generate confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
# for the tunning
```

```python
Y_pred_prob = classifier.predict_proba(X_test)
print(Y_pred_prob)
```

```python
Y_pred_class=[]
for value in Y_pred_prob[:,1]:
    if value > 0.42: #after tunning select proper thresold values.
        Y_pred_class.append(1)
    else:
        Y_pred_class.append(0)
print(Y_pred_class)
```

```python
# GEnerate confusion matrix of logistic regression

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred_class)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred_class))

acc=accuracy_score(Y_test, Y_pred_class)
print("Accuracy of the model: ",acc)
```

```python

#find the error for different thresold

for a in np.arange(0.3,0.61,0.01):
    predict_mine = np.where(Y_pred_prob[:,1] > a, 1, 0)
    cfm=confusion_matrix(Y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :",
          cfm[1,0]," , type 1 error:", cfm[0,1])
```
